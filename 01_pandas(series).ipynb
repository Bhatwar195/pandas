### Pandas (Series)
Pandas Series is a one-dimensional labeled data structure used to store and manipulate data efficiently. It supports various data types and provides powerful indexing capabilities, making it a fundamental building block for data analysis in Python.


```python
import pandas as pd
```

```python
# check version
pd.__version__
```

```python
# Creating a Python list with mixed data types
list1 = [1, 2, 3, 3.14, "Python", "ML", "Ds"]
print(list1)
```

```python
# Converting List to Pandas Series
series1 = pd.Series(list1)
print(series1)
```

```python
# Checking the data type of Pandas Series
type(series1)
```

```python
# Creating an Empty Series
series2 = pd.Series([])
print(series2)
```

### Creating Series using a list

```python
series3 = pd.Series([1, 2, 3, 4, 5, 6, 7])
print(series3)
print(type(series3))
```

### Creating Series with custom index

```python
series3 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'])
print(series3)
```

### Creating Series with custom index and data type

```python
series3 = pd.Series([1, 2, 3, 4], index=['a', 'b', 'c', 'd'], dtype=float)
print(series3)
```

### Conclusion
- Pandas Series is a one-dimensional labeled data structure.
- It can store mixed data types.
- Index and data type can be customized.
- Series is the foundation for DataFrames.
