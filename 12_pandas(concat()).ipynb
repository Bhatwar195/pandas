## concate() Function 
**`concat()`** is a Pandas function used to combine multiple DataFrames or Series along rows or columns.

```python
import pandas as pd
```

### Creating Sample series 

```python
sr1=pd.Series([1,2,3,4,5,6,7])
```

```python
sr2=pd.Series([8,9,10,11])
```

### Concatenate series 

```python
pd.concat([sr1,sr2])
```

### Creating Sample DataFrames

```python
df1 = pd.DataFrame({
    "Name": ["Amit", "Rahul"],
    "Age": [21, 22]
})
df1
```

```python
df2 = pd.DataFrame({
    "Name": ["Neha", "Priya"],
    "Age": [20, 23]
})
df2
```

### Concatenate Rows (axis=0)

```python
df_row = pd.concat([df1, df2])
df_row
```

### Concatenate Columns (axis=1)

```python
pd.concat([df1,df2],axis=1)
```

### Reset Index after Concatenation

```python
df_row_reset = pd.concat([df1, df2], ignore_index=True)
df_row_reset
```

### Concatenate with join  

```python
pd.concat([df1,df2],axis=1,join='inner')
```

```python
pd.concat([df1,df2],axis=1,join='outer')
```

### Using Keys

```python
pd.concat([df1, df2], keys = ['df1','df2'],axis=1)
```

```python
pd.concat([df1, df2], keys = ['First df','second df'],axis=1)
```

```python
df3 = pd.DataFrame({'Marks': [40, 63, 91, 34]})
df3
```

```python
pd.concat([df1,df3],axis=1)
```

```python
pd.concat([df1,df3],axis=0)
```

### Conclusion
The concat() function is used to combine DataFrames along rows or columns.
