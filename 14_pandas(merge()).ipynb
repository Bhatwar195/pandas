### merge() function 
- **`merge()`** in Pandas is used to combine DataFrames based on common column values.

```python
import pandas as pd 
```

### Creating Sample DataFrames

```python
df1 = pd.DataFrame({
    "ID": [1, 2, 3],
    "Name": ["Amit", "Rahul", "Neha"]
})
```

```python
df2 = pd.DataFrame({
    "ID": [1, 2, 4],
    "Marks": [85, 78, 90]
})
```

```python
display(df1,df2)
```

### Inner Merge

```python
df_inner = pd.merge(df1, df2, on="ID", how="inner")
df_inner
```

### Left Merge

```python
df_left = pd.merge(df1, df2, on="ID", how="left")
df_left
```

### Right Merge

```python
df_right = pd.merge(df1, df2, on="ID", how="right")
df_right
```

### Outer Merge

```python
df_outer = pd.merge(df1, df2, on="ID", how="outer")
df_outer
```

### Merge with Different Column Names

```python
df3 = pd.DataFrame({
    "Student_ID": [1, 2, 3],
    "City": ["Nagpur", "Pune", "Mumbai"]
})
```

```python
df_merge_diff = pd.merge(df1, df3, left_on="ID", right_on="Student_ID")
df_merge_diff
```

```python
pd.merge(df1,df2,on='Id')
```

### Outer merge with indicator 

```python
pd.merge(df1,df2,how='outer',indicator=True)
```

### Merge with suffixes 

```python
df1.merge(df2,left_index=True,right_index=True,suffixes=('_High','_Low'))
```
