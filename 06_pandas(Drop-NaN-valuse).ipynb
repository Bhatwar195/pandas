## Pandas Day - 6 (Drop null values)

NaN (Not a Number) represents missing or undefined values in a dataset.

#### *Checks where NaN values are present-
**isnull() / isna():**  
#### *Drop the NaN values.    
**dropna() parameters:**
- axis: drop rows (0) or columns (1)
- how: "any" or "all"
- subset: specific columns to check

```python
import pandas as pd 
import numpy as np
```

```python
df=pd.read_csv('Fortune_10.csv')
df
```

### Check null value using isna() , isnull()

```python
df.isna()
```

```python
df.isnull()
```

### Checking the number of null values using sum().

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

### Checking Null value using notnull() & notna().

```python
df.notnull()
```

```python
df.notna()
```

### Chacking Sum fo not null values .

```python
df.notna().sum()
```

```python
df.notna().sum().sum()
```

### Drop the null Values using Dropna()

```python
df=pd.read_csv("Fortune_10 - Copy.csv",sep="\t")
df
```

```python
df.dropna()
```

### Drop null values column

```python
df.dropna(axis=1)
```

```python
df.dropna(how='any')
```

```python
df.dropna(how='all')
```

```python
df.dropna(thresh=2)
```

```python
df.dropna(subset=['Growth'])
```

```python
df.dropna(inplace=True)
```

```python
df.head()
```
