### pivot_table()
- **pivot_table()** is used to summarize and aggregate data in a table format.

```python
import pandas as pd 
```

### Creating Sample Data

```python
df = pd.DataFrame({
    "Name": ["Amit", "Amit", "Rahul", "Rahul"],
    "Subject": ["Maths", "Science", "Maths", "Science"],
    "Marks": [85, 88, 78, 75]
})
```

```python
df
```

### Basic pivot_table()

```python
pd.pivot_table(df, values="Marks", index="Name")
```

### Using Aggregation Function

```python
pd.pivot_table(df, values="Marks", index="Name", aggfunc="mean")
```

### Multiple Aggregations

```python
pd.pivot_table(df, values="Marks", index="Name", aggfunc=["mean", "max"])
```
