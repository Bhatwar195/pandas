### Dataframe Creation 

A DataFrame is a two-dimensional data structure in Pandas that stores data in rows and columns.

```python
import pandas as pd
```

### Creating empty data frame

```python
df=pd.DataFrame([])
print(df)
```

### Creating dataframe from list

```python
list1=['a','e','i','o','u']
df1=pd.DataFrame(list1)
df1
```

### Creating dataframe from list of list

```python
list2=[[1,2,3],[4,5,6],[7,8,9]]
df2=pd.DataFrame(list2)
df2
```

### Creating dataframe from dict or dictionary or python dictionary

```python
data = {
    "Name": ["Amit", "Rahul", "Neha", "Priya"],
    "Age": [21, 22, 20, 23],
    "Marks": [85, 78, 90, 88]
}

df = pd.DataFrame(data)
df
```
