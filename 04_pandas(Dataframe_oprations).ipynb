## Pandas DataFrame Operations

```python
import pandas as pd
```

```python
# Creating DataFrame
data = {
    "Name": ["Amit", "Rahul", "Neha", "Priya"],
    "Age": [21, 22, 20, 23],
    "Marks": [85, 78, 90, 88]
}

df = pd.DataFrame(data)
df
```

### Checking Data

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

### Column Operations

```python
df["Age"]
```

```python
df[["Name", "Marks"]]
```

### Adding a New Column

```python
df["Passed"] = df["Marks"] >= 80
df
```

### Basic Calculations

```python
df["Marks"].mean()
```

```python
df["Marks"].max()
```

```python
df["Marks"].min()
```

### Filtering Data

```python
df[df["Marks"] > 80]
```
