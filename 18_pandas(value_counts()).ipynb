## Value counts()
- **value_counts()** is used to count unique values in a column.

### Creating Sample Data

```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    "City": ["Nagpur", "Pune", "Nagpur", "Mumbai", "Pune", "Nagpur"],
    "Gender": ["M", "F", "F", "M", "M", "M"]
})
```

### value_counts() on Single Column

```python
df["City"].value_counts()
```

### value_counts() with Normalization

```python
df["City"].value_counts(normalize=True)
```

### value_counts() Including NaN

```python
df["City"].value_counts(dropna=False)
```

### value_counts() on Multiple Columns

```python
df.value_counts(["City", "Gender"])
```
