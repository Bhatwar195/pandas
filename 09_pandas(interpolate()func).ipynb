## Pandas Day -9 ( Handling Missing Values using interpolate() )

### Pandas Interpolate ()
----------------------------------
Pnadas interpolate() function is basically used to fill NaN values in DataFrame and Series.

 - Very Powerfull Function
 - Uses various interpolation techniques

Syntax: 
    DataFrame.interpolate()
    Series.interpolate()
    
Parameters :  
- method  :: inplace  
- axis    :: limit_direction  
- limit   :: limit_area  

```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("student_10_day13.csv")
```

```python
df.head()
```

### Linear Interpolation

```python
df_linear = df.interpolate()
df_linear
```

```python
df.interpolate(method='linear')
```

### Interpolation with Method 

```python
df.interpolate(method='bfill')
```

```python
df=pd.read_csv("student_10_day13.csv",parse_dates=['Date'],index_col=['Date'])
```

```python
df.interpolate(method='time')
```

### Interpolation with index method 

```python
df=pd.read_csv("student_10_day13.csv",index_col='Class')
```

```python
df.interpolate(method='index')
```

### Interpolation with Method 'polynomial'

```python
df=pd.read_csv("student_10_day13.csv")
```

```python
df.interpolate(method='polynomial',order=3)
```

### Interpolation with Method 'nearest'

```python
df.interpolate(method='nearest')
```

### Interpolation with Limit

```python
df.interpolate(limit=2)
```

### Interpolation with Direction

```python
df.interpolate(limit_direction='backward')
```

```python
df.interpolate(limit_direction='both')
```

```python
df.interpolate(limit_area='inside')
```
