### Parameters in read_csv

```python
df=pd.read_csv("Fortune_10.csv",header=1)# header is use for to make headings 
df
```

```python
df=pd.read_csv("Fortune_10.csv",header=None)
df
```

```python
df=pd.read_csv("Fortune_10.csv",names=['A','B','C','D','E','F','G','H','I','J'])# extra index print nan column 
df
```

### Reading Limited Rows (nrows)

```python
df=pd.read_csv("Fortune_10.csv",nrows=7)
df
```

### Skipping Rows (skiprows)

```python
df = pd.read_csv("Fortune_10.csv", skiprows=1)
df
```

```python
df=pd.read_csv("Fortune_10.csv",usecols=[4,5,6])
df
```

```python
df=pd.read_csv("Fortune_10.csv",index_col=1)
df
```

### Using Separator (sep)

```python
df = pd.read_csv("student_result1.csv", sep=",")
df
```
