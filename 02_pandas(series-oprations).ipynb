### Pandas Series â€“ Creation and Operations

```python
import pandas as pd
```

```python
# Creating a Series with custom index, name, and data type
series = pd.Series(
    74,
    index=['a', 'b', 'c', 'd', 'e'],
    name="Similar series",
    dtype=float
)
series
```

### Creating Series from a Python Dictionary

```python
data = {"a": 12, "b": 34, "c": 56}
x = pd.Series(data)
x
```

### Accessing Elements from a Series

```python
series = pd.Series([1, 2, 3, 4, 5, 6, 7, 8])
series
```

```python
# Slicing the series
series[2:7]
```

### Arithmetic Operations on Series

```python
s1 = pd.Series([1, 2, 3, 4, 5])
s2 = pd.Series([10, 20, 30, 40, 50])
```

```python
# Addition
print(s1 + s2)
```

```python
# Multiplication
print(s1 * s2)
```

```python
# Power operation
print(s1 ** 3)
```

### Statistical Operations

```python
mean_value = pd.Series([1, 2, 3, 4, 5, 6, 7, 8]).mean()
mean_value
```

```python
# Maximum value
max(s1)
```

```python
# Minimum value
min(s2)
```

### Conditional Filtering

```python
a = pd.Series([3, 6, 9, 12, 15])
a[a < 9]
```

```python
s1[s1 < 2]
```

### Dropping Elements from Series

```python
b = s1.drop(3)  # Drops element at index 3
b
```

```python
# Adding two series
a.add(b)
```

### Conclusion
- Pandas Series supports arithmetic, statistical, and conditional operations.
- Index-based operations make data manipulation flexible.
- Series methods like `mean()`, `drop()`, and `add()` simplify analysis.
