###  groupby() function  
- Pandas GroupBy function is used to split the data into groups based on some criteria.

```python
import pandas as pd 
```

```python
df=pd.read_csv("student_result1.csv")
df
```

### Grouping Data by Single Column

```python
# method 1
group_section=df.groupby(by='Section')
group_section.groups
```

```python
# method 2
df.groupby(by='Section').groups
```

### Grouping by Multiple Columns

```python
df.groupby(['Section','Class']).groups
```

```python
for Class, df in group_section:
 print(Class)
 print(df)
```

```python
for Section , df in group_section:
 print(Class)
 print(df)
```

### A single group can be selected using get_group():

```python
df=pd.read_csv("student_result1.csv")
```

```python
g2 = df.groupby('Class').get_group(10)
g2
```

```python
g3 = df.groupby('Section').get_group('A')
g3
```

### Applying Aggregation (mean)

```python
group_section.mean()
```

### Grouping with sum

```python
group_section.sum()
```

### Multiple Aggregations

```python
g1.agg(['mean','max','sum'])
```

The groupby() function is used to group data and apply aggregation operations.
