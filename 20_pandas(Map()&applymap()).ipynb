##  map() and applymap()
- **map()** is used for Series, while **applymap()** is used for element-wise operations on DataFrames.

```python
import pandas as pd 
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame({
    "Name": ["Amit", "Rahul", "Neha", "Priya"],
    "City": ["nagpur", "pune", "mumbai", "nagpur"],
    "Marks": [85, 78, 90, 88]
})
```

### map() on a Series

```python
df["City"] = df["City"].map(str.upper)
df
```

### map() using Dictionary

```python
grade_map = {85: "A", 78: "B", 90: "A+", 88: "A"}
df["Grade"] = df["Marks"].map(grade_map)
df
```

### applymap() on DataFrame

```python
df_applymap = df[["Marks"]].applymap(lambda x: x + 2)
df_applymap
```
