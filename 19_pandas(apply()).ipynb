## apply() function 
- **apply()** is used to apply a function to rows or columns of a DataFrame.

```python
import pandas as pd 
```

### Creating Sample DataFrame

```python
df = pd.DataFrame({
    "Name": ["Amit", "Bhavesh", "Pranay", "Priya"],
    "Marks": [85, 78, 90, 88]
})
```

### apply() on a Column

```python
df["Marks_plus_5"] = df["Marks"].apply(lambda x: x + 5)
df
```

### apply() with Custom Function

```python
def grade(marks):
    if marks >= 85:
        return "A"
    elif marks >= 75:
        return "B"
    else:
        return "C"
```

```python
df["Grade"] = df["Marks"].apply(grade)
df
```

### apply() on Rows (axis=1)

```python
df["Result"] = df.apply(
    lambda row: "Pass" if row["Marks"] >= 80 else "Fail",
    axis=1
)
df
```

### apply() on Columns (axis=0)

```python
df[["Marks"]].apply("mean")
```
