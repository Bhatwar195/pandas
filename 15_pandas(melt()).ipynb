### melt() Function 
- **`melt()`** in Pandas is used to convert data from wide format to long format.

### Creating Sample DataFrame

```python
import pandas as pd 
```

```python
df = pd.DataFrame({
    "Name": ["Amit", "Rahul", "Neha"],
    "Maths": [85, 78, 90],
    "Science": [88, 75, 92]
})
```

```python
df
```

### Basic melt()

```python
df_melt = pd.melt(df)
df_melt
```

### Using id_vars

```python
df_melt_id = pd.melt(df, id_vars=["Name"])
df_melt_id
```

### Using value_vars

```python
df_melt_val = pd.melt(df, id_vars=["Name"], value_vars=["Maths"])
df_melt_val
```

### Renaming Columns

```python
df_melt_rename = pd.melt(
    df,
    id_vars=["Name"],
    var_name="Subject",
    value_name="Marks"
)
df_melt_rename
```
