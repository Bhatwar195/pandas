### Pandas sort_values() & sort_index()
- **sort_values()** sorts data by column values, while sort_index() sorts by index.

```python
import pandas as pd 
```

### Creating Sample DataFrame

```python
df = pd.DataFrame({
    "Name": ["Amit", "Rahul", "Neha", "Priya"],
    "Age": [21, 22, 20, 23],
    "Marks": [85, 78, 90, 88]
})
```

```python
df
```

### Sort by Single Column

```python
df.sort_values("Marks")
```

### Sort in Descending Order

```python
df.sort_values("Marks", ascending=False)
```

### Sort by Multiple Columns

```python
df.sort_values(["Age", "Marks"])
```

### Sort Index

```python
df.sort_index()
```

### Sort Index in Descending Order

```python
df.sort_index(ascending=False)
```
